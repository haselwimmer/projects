# McKinley Mine - Planetscope data analysis

### Scope
- Offline analysis of Planetscope scenes

### Pseudocode for area average NDVI or other band ratios
1. Open VMU boundaries vectors
2. Build list of raster datasets; e.g. for Planetscope query subdirectories or for SentinelHub perform API query
3. Create empty list to store results for all VMU's 
4. Loop over VMU's:
   - Loop over raster datasets:
     - For each raster dataset calculate average NDVI for VMU extent
     - Populate into dataframe
5. Concatenate the list into a single DF for all VMU's and raster datasets

```python
# Import libraries
import os
from glob import glob
import matplotlib.pyplot as plt

import rioxarray as rxr
import pandas as pd
import geopandas as gpd
import rasterio
import rasterstats as rs
import numpy as np
import seaborn as sns

# Setting consistent plotting style throughout notebook
sns.set_style("white")
sns.set(font_scale=1.5)
```

### Functions

```python
def ndvi_calc(vis, nir):
    """Calculate NDVI (Normalized Difference Vegetation Index) from visible and near-infared bands

    Parameters
    -----------
    vis : xarray DataArray
        visible raster
    nir : xarray DataArray
        near-infrared raster

    Returns
    -----------
    output : list
        A three item list containing string of band math algorithm name and description, and xarray of the result

    """
    name = "NDVI"
    desc = "Normalized Difference Vegetation Index"
    result = (nir-vis) / (nir+vis)
    
    return [name, desc, result]

def savi_calc(vis, nir, L=0.5):
    """Calculate SAVI (Soil Adjusted Vegetation Index) from visible and near-infared bands

    Parameters
    -----------
    vis : xarray DataArray
        visible raster
    nir: xarray DataArray
        near-infrared raster
    L : float
        soil adjustment constant

    Returns
    -----------
    output : list
        A three item list containing string of band math algorithm name and description, and xarray of the result
    """
    name = "SAVI"
    desc = "Soil Adjusted Vegetation Index"
    result = ((nir - vis) / (nir + vis + L)) * (1 + L)
    return [name, result]
```

### Import vectors

```python
# Open vector file
vmu_path = r"E:\McKinley\vectors\VMU_OSM_boundaries_NAD83_shifted.shp"
vmu_poly = gpd.read_file(vmu_path)

# Plot the data
# f, ax1 = plt.subplots(figsize=(12, 6))
# vmu_poly.plot(ax=ax1)
# plt.show()
```

### Planetscope data - perform band math for VMU's

```python
# Get paths to Planetscope directories
dirs = glob('E:\McKinley\planetscope\*\*McKinley*')

# Create empty list to store NDVI results
output_list = []

# Loop over each Planetscope directory
for dir in dirs:

    # Extract the date from the directory name
    path_components = dir.split(os.sep)
    dir_name = path_components[-1]
    dir_name_components = dir_name.split('_')
    date = dir_name_components[-1]

    # Get paths to bands to be used for NDVI calculation
    band_paths = sorted(glob(os.path.join(dir, "*B*[3-4].tif")))

    # Open the first band and read the CRS
    band = rxr.open_rasterio(band_paths[0], masked=True).squeeze()
    planetscope_crs = band.rio.crs

    # Clean up memory
    band = None

    # Ensure the vector and raster are in the same CRS
    vmu_poly = vmu_poly.to_crs(planetscope_crs)

    # Open bands and clip to extent of poly
    band3 = rxr.open_rasterio(band_paths[0], masked=True).rio.clip(vmu_poly.geometry,
                                                                   from_disk=True).squeeze()

    band4 = rxr.open_rasterio(band_paths[1], masked=True).rio.clip(vmu_poly.geometry,
                                                                   from_disk=True).squeeze()

    # Perform band math
    #output = savi_calc(band3, band4, L=1)
    output = ndvi_calc(band3, band4)
    output_xr = output[2]
    bandmath_name = output[1]

    # Clean up memory
    band3 = None
    band4 = None

    
    # Save the band math result to disk
    output_file = date + "_" + bandmath_name + ".tif"
    output_path = os.path.join(dir, output_file)
    output_xr.rio.to_raster(output_path, dtype="float32")

#     # get the metadata of original GeoTIFF:
#     meta = output_xr.meta

#     # set the source metadata as kwargs we'll use to write the new data:
#     kwargs = meta

#     # update the 'dtype' value to uint16:
#     kwargs.update(dtype='float32')

#     # Write band math results to a new GeoTIFF file
#     with rasterio.open(output_path, 'w', **kwargs) as dst:
#         dst.write_band(output_xr, bandmath_name)

        
    # Use zonal stats to calculate mean NDVI for each VMU
    output_vmu = rs.zonal_stats(vmu_poly,
                                output_xr.values,
                                affine=output_xr.rio.transform(),
                                geojson_out=True,  # Optional if you want to plot
                                copy_properties=True,  # Optional if you want to plot
                                stats=["mean", "max"])

    # Clean up memory
    output_xr = None

    # Convert geojson results to a GDF
    output_vmu_gdf = gpd.GeoDataFrame.from_features(output_vmu)

    # Rename columns
    output_vmu_gdf.rename(columns={"max": "bandmath_max",
                                   "mean": "bandmath_mean"},
                          inplace=True)

    # Add date column to VMU df
    output_vmu_gdf['date'] = date

    # Append values to the list
    output_list.append(output_vmu_gdf)

# Concatenate all the dates into a single DF
output_vmu_all_dates = pd.concat(output_list, axis=0)

# Reformat the date
output_vmu_all_dates['date'] = pd.to_datetime(output_vmu_all_dates['date'])

# Reorder by date
output_vmu_all_dates.set_index("date", inplace=True)
```

### Plot results

```python
# Remove 999
output_vmu_all_dates = output_vmu_all_dates[output_vmu_all_dates.VMU != 999]

# Setup plot
fig, ax = plt.subplots(figsize=(15, 15))

# Build list of unique sites for legend
sites = output_vmu_all_dates.VMU.unique()

# Convert the gdf to pd in order to plot properly
output_vmu_all_dates = pd.DataFrame(output_vmu_all_dates)

# Group the df by site for plotting purposes
output_vmu_all_dates.reset_index().groupby('VMU').plot(x='date', y='bandmath_mean', ax=ax)

# Legend and labels
plt.legend(sites)
ax.set(title="Band math results for McKinley VMU's",
       xlabel="Date",
       ylabel=bandmath_name)

plt.show()
```
